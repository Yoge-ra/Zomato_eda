```python
# looking for outliers

fig,axes = plt.subplots(1,2, figsize=(15,10)) # Creating 2x2 grid for box plots.

# box plot for average cost for two
sns.boxplot(ax = axes[0],data=final_df,x=final_df['Average Cost for two'])
axes[0].set_title('Average Cost for Two')

# box plot for price range
sns.boxplot(ax = axes[1],data=final_df,y='Price range')
axes[1].set_title('Price Range')

# To prevent overlap
plt.tight_layout()

plt.show()
```

Price range and Average Cost of two are important columns so, we must remove the outliers

```python
print(final_df['Price range'].max())
print(final_df['Average Cost for two'].max())
```

##### By IQR technique

```python
# function to remove outliers
def iqr_tech(Column):
  Q1 = np.percentile(Column,25)
  Q3 = np.percentile(Column,75)

  IQR = Q3-Q1

  lower_fence = Q1 - (1.5*IQR)
  upper_fence = Q3 + (1.5*IQR)

  return lower_fence,upper_fence
```

```python
lower_bound,upper_bound = iqr_tech(final_df['Average Cost for two'])

final_df = final_df[(final_df['Average Cost for two'] >= lower_bound) & (final_df['Average Cost for two'] <= upper_bound)]
```

```python
lower_bound,upper_bound = iqr_tech(final_df['Price range'])

final_df = final_df[(final_df['Price range'] >= lower_bound) & (final_df['Price range'] <= upper_bound)]
```

**4. Are there any country wise variation in cuisine preference?**

```python
# Grouping the data by country and cuisines
country_cuisine = final_df.groupby(['Country','Cuisines']).size().reset_index(name='Count')
# Determine the top cuisines in each country
top_cuisines_by_country = country_cuisine.groupby('Country').apply(lambda x: x.nlargest(5,"Count")).reset_index(drop=True)

# Visualize top cuisines in each country
for country in top_cuisines_by_country['Country'].unique():
  plt.figure(figsize=(10,5))
  dat = top_cuisines_by_country[top_cuisines_by_country['Country']==country]
  sns.barplot(x=dat['Cuisines'],y=dat['Count'],data=dat,palette='viridis')
  plt.title(f'Top cuisines in {country}')
  plt.xlabel('Cuisines')
  plt.ylabel("Count")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()
```

**5. Is there a correlation between average ratings and the number of votes?**
